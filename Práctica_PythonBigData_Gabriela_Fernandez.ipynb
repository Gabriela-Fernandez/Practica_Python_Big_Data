```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precios. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
def suma_total(iterable):
    return reduce(lambda x, y: x + y[1], iterable.items(), 0)
suma_total(articulos)
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%
def total_descuento(iterable, descuento):
    return reduce(lambda x, y: x + ((1-(descuento/100))*y[1]), iterable.items(), 0)
total_descuento(articulos, 10)
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
def total_mayor_100(iterable, descuento):
    if suma_total(iterable) > 100:
        return total_descuento(iterable, descuento)
    else:
        return suma_total(iterable)
total_mayor_100(articulos, 10)
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)
def total_mas_iva(iterable, iva=0.21):
    return reduce(lambda x, y: x + ((1+iva)*y[1]), iterable.items(), 0)
total_mas_iva(articulos)
```

```python
# E.lista los artículos cuyo precio es superior a 20€
mayores_20 = list(filter(lambda precio: precio > 20, articulos.values()))
mayores_20
```
